# Python Literals

## Overview

A **Literal** is a raw, fixed data value explicitly written into the source code of a program. It is *not* a variable, nor is it the result of an expression or a function call. It is the raw data itself.

### Identifying Literals

```python
x = 10        # '10' is an integer literal. 'x' is a variable.
y = 5         # '5' is an integer literal.
z = x + y     # 'x + y' is an expression, NOT a literal.
name = input()# The value comes from runtime input, NOT a literal.

```

*Note:* In computer science terminology, literals are often referred to as "Constants," though Python does not have strict constant variables in the way languages like C++ (`const`) or Java (`final`) do.

---

## 1. Integer Literals

These are whole numbers (positive, negative, or zero) without a decimal point.

### The Underscore Feature (Python 3.6+)

When dealing with large numbers, it is difficult to read them quickly (e.g., `1000000000`). Python allows you to use underscores `_` as visual separators.

* The Python interpreter completely ignores the underscores. They exist purely for developer readability.

```python
# Standard Literal
population = 1500000

# Separated Literal (Much easier to read)
population = 1_500_000

# You can group them however you want (e.g., Indian numbering system)
rupees = 1_50_000 # 1 Lakh, 50 Thousand

```

**Constraints:** Underscores must be *between* digits. They cannot appear at the very beginning (`_100` is read as a variable name), the very end (`100_`), or immediately next to a decimal point.

---

## 2. Float Literals

These are numbers that contain a decimal point or use scientific notation.

```python
# Standard Float
pi = 3.14159

# Float via decimal enforcement (Even if it's a whole number)
weight = 15.0  

# Scientific Notation (2 * 10^2)
scientific_a = 2e2   # Evaluates to 200.0

# Scientific Notation with negative exponent (1.25 * 10^-1)
scientific_b = 1.25e-1 # Evaluates to 0.125

```

*Note:* The underscore separator feature works for floats as well (e.g., `3_141.59`).

---

## 3. Boolean Literals

Python has exactly two boolean literals, representing truth values.
**Constraint:** They must be capitalized.

```python
is_active = True
has_errors = False

# ❌ invalid_bool = true  (Raises NameError: name 'true' is not defined)

```

---

## 4. Complex Literals

Python natively supports complex numbers (numbers with a real and an imaginary part). Python uses `j` or `J` to denote the imaginary unit (in mathematics, this is usually `i`).

```python
# Real part is 4, Imaginary part is 5
impedance = 4 + 5j 

# Can be combined with floats
signal = 1.5 + 2.5j

# Can be combined with underscores
large_signal = 1_000 + 5_000j

```

---

## 5. String Literals

Strings represent text data. Python is highly flexible with how you define string literals, allowing you to avoid "escaping" quotes if you use the opposite quote type.

### A. Single Quotes `''`

```python
name = 'Alice'

```

### B. Double Quotes `""`

Functionally identical to single quotes. They are incredibly useful if your string contains an apostrophe.

```python
# Using double quotes prevents the apostrophe from breaking the string
message = "It's a beautiful day" 

# Alternatively, nesting quotes
quote = 'He said, "Hello there!"'

```

### C. Triple Quotes `''' '''` or `""" """`

Used for multi-line strings or Docstrings (documentation built into the code).

```python
query = """
SELECT id, name
FROM users
WHERE active = 1;
"""

```

| Literal Type | Examples | Valid Syntax Rules |
| --- | --- | --- |
| **Integer** | `42`, `-7`, `1_000` | No decimals. `_` must be between digits. |
| **Float** | `3.14`, `10.0`, `2e3` | Must have decimal or `e`/`E`. |
| **Boolean** | `True`, `False` | Must be Capitalized. |
| **Complex** | `3+4j`, `1.5j` | Must use `j` or `J`. |
| **String** | `'A'`, `"Text"`, `"""Doc"""` | Matching quote pairs required. |

# Python Integer Literals and Number Systems

While we interact with numbers in base-10 (Decimal) in our everyday lives, computer architecture fundamentally operates in base-2 (Binary).

Python allows you to write integer literals directly in **Binary**, **Octal**, and **Hexadecimal** formats. This is incredibly useful when working with low-level systems programming, networking protocols, cryptography, or bitwise operations, where the exact bit-pattern of a number is more important than its decimal value.

Regardless of which base you use to write the literal, **Python stores it as a standard integer**. If you print it, Python will output the base-10 decimal equivalent by default.

---

## 1. Decimal (Base-10)

This is the standard numbering system using digits `0-9`. No prefix is required.

```python
num = 17
print(num) # Output: 17

```

---

## 2. Binary (Base-2)

Binary uses only two digits: `0` and `1`.

* **Prefix:** `0b` or `0B` (Zero followed by 'b').

### Engineering Example

Binary is often used for bitmasking (managing system flags).

```python
# The decimal number 17 is 10001 in binary
# (16 + 0 + 0 + 0 + 1 = 17)

binary_literal = 0b10001
print(binary_literal) 
# Output: 17

```

---

## 3. Octal (Base-8)

Octal uses digits `0-7`. It was historically useful because one octal digit maps exactly to three binary digits (bits), making it a shorthand for binary.

* **Prefix:** `0o` or `0O` (Zero followed by the letter 'o').

### Engineering Example

Octal is heavily used in Unix/Linux systems for file permissions (e.g., `chmod 755`).

```python
# The decimal number 17 is 21 in octal
# (2 * 8) + (1 * 1) = 17

octal_literal = 0o21
print(octal_literal) 
# Output: 17

# File permission example: rwxr-xr-x
permissions = 0o755 

```

---

## 4. Hexadecimal (Base-16)

Hexadecimal (or "Hex") uses digits `0-9` and letters `A-F` (representing 10-15). It is the most common shorthand for binary today because one hex digit maps exactly to four bits (a nibble), and two hex digits map exactly to one byte (8 bits).

* **Prefix:** `0x` or `0X` (Zero followed by 'x').

### Engineering Example

Hex is universally used for memory addresses, cryptographic hashes, and web color codes.

```python
# The decimal number 17 is 11 in hexadecimal
# (1 * 16) + (1 * 1) = 17

hex_literal = 0x11
print(hex_literal) 
# Output: 17

# Hex color code example (White)
white_color = 0xFFFFFF

```

---

## 5. Constraints and Usage Rules

1. **Integers Only:** These prefixes (`0b`, `0o`, `0x`) can **only** be used for integers. You cannot write a floating-point number like `0b10.11`.
2. **Complex Numbers:** You can use these base literals for the **Real** part of a complex number, but **not** the Imaginary part.
```python
# Valid: Binary real part, decimal imaginary part
c1 = 0b1010 + 5j  # 10 + 5j

# ❌ Invalid: Cannot use base prefix for the imaginary part
# c2 = 10 + 0b101j # SyntaxError

```
